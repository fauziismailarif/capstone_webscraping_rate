## Requesting the Data and Creating a BeautifulSoup

selanjutnya kita lakukan impor dari web exchange (https://www.exchange-rates.org/history/IDR/USD/T)

```python
import requests

url_get = requests.get('https://www.exchange-rates.org/history/IDR/USD/T')
```

untuk memvisualisasikan apa yang telah kita dapat dari `request.get`, dapat kita lihat pada data berikut:


```python
url_get.content[1:500]
```

Untuk melihat lebih jelas atas kode diatas, kita menggunakan BeautifulSoup agar lebih mudah dibaca.

```python
from bs4 import BeautifulSoup 

soup = BeautifulSoup(url_get.content,"html.parser")
```

Penggunaan BeautifulSoup akan menampilkan data seperti ini:

```python
soup
```

Untuk mencari kata kunci, kita menggunakan `.find()` 

```python
table = soup.find('table')
print(table.prettify()[1:500])
```

Karena data sumber berupa tabel dan lebih mudah jika menggunakan pendekatan parsing table. Maka kita menggunakan mekanisme python BeautifulSoup parsing table https://stackoverflow.com/questions/23377533/python-beautifulsoup-parsing-table

```python
# parsing tabel dan masukkan datanya dalam sebuah list bernama data
data = []
table = soup.find('table')
table_body = table.find('tbody')

rows = table_body.find_all('tr')
for row in rows:
	cols = row.find_all('td')
	cols = [el.text.strip() for el in cols]
	data.append([el for el in cols])
```

## Creating data frame & Data wrangling

Setelah parsing tabel dilakukan, maka kita buat dataframe

```python
import pandas as pd

dataawal = pd.DataFrame(data, columns=['date', 'day', 'rate', 'description'])
dataawal.head()
```

kolom description tidak kita gunakan, sehingga dapat didrop.

```python
dataolah = dataawal.drop('description',1)
```

Selanjutnya kita ubah tipe kolomnya menjadi lebih sesuai dengan data yang ada.

```python
dataolah.rate = dataolah.rate.str.replace(",", "")
dataolah.rate = dataolah.rate.str.replace(" IDR", "")
dataolah.rate = dataolah.rate.astype('float64')
dataolah.date = dataolah.date.astype('datetime64')
dataolah.day = dataolah.day.astype('category')
```

Menggunakan *set_index* untuk meletakkan kolom *date* sebagai index

```python
dataolah = dataolah.set_index('date')
```

Agar data berurutan mulai dari awal, kita menggunakan **[::-1]**

```python
dataolah = dataolah[::-1]
dataolah[:10]
```
